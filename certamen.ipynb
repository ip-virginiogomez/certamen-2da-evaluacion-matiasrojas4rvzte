### Ejercicio 3.2 – Clasificador de notas (16 pts)

Escribe un programa que:
1. Pida una **nota entre 1.0 y 7.0**.
2. Valide el rango → si no es válida, muestra `"Nota inválida."` y termina.
3. Muestre el **concepto** y si está **aprobado o reprobado**.

| Rango        | Concepto      |
|--------------|---------------|
| 6.0 – 7.0    | Excelente     |
| 5.0 – 5.9    | Bueno         |
| 4.0 – 4.9    | Suficiente    |
| 1.0 – 3.9    | Insuficiente  |

**Salida esperada:**
```
Ingresa tu nota (1.0 - 7.0): 5.5
Concepto: Bueno
Estado: Aprobado
```


````python
# Ejercicio 3.2 – Clasificador de notas
# Escribe tu código aquí:
Nota=float(input("ingrese su nota"))
if Nota >=6.0 :
    print("Exelente,estas aprobado")
elif Nota>=5.0 :
    print("Bueno,estas aprobado")
elif Nota>=4.0 :
    print("Suficiente,aprobado")
elif Nota<=3.9 :
    print("Insuficiente,desaprobado")
else:
    print("nota invalida")
````
